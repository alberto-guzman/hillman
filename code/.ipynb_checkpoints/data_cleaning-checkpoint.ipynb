### Data Overview 

### Alumni Tracker:

-	Contains names and most info on all alumni who have completed the Hillman Academy. 
-	Basic and contact info, participation years, college, graduation status, public profiles, publications, stipend/grant paid , awards.
-	Info in Green is the cleaned up information. The info afterwards are the different sources of data that go into making the green data.
-	When collecting data and a conflict emerges priority is given to Direct Student Response> National student Clearinghouse> Public Profile> Old data point
-	UPCI Partcipation year denotes which year a student participated. Subsequent years are indicated with a “/”. If a student had multiple years Funding and PI will have a “/” to separate years if there were different funding or mentors. 


### Appplications: 

-	Application data for all years. 
-	Applications changed multiple times over the years but information is mostly the same. 
-	Basic info, application form, transcripts, essays, letters of recommendation. (not all available)
-	Data includes; demographics, academic records, stipend eligibility. 


### Things to Note: 

-	Started 2009. No sites yet.
-	CURE Grant started 2010
-	DDCF Grant started 2012
-	2 Year YES program started in 2019
-	2020 is a virtual program and has no sites.
-	In 2010 Sites Cancer Biology (CB or Hillman) and CoSBBI are created.
-	In 2011 WCRC site is added
-	In 2012 Tumor immunology (TI/ICI) is added changed to ICI in 2019
-	In 2013 DiscoBio is added, changed to CompBio in 2019
-	In 2015 CEBIG Site is Added, Changed to TDX in 2019
-	Sites that are not among those above are when students worked with a mentor outside the normal labs of the program.

```r
library(tidyverse)
library(readxl)
library(magrittr)
library(janitor)
```

```r
alum <- read_excel("/Users/albertoguzman-alvarez/Box/Data/Analysis/raw_data/Alumni Tracker SL 6.15.2020 copy.xlsx", sheet = 1, range = cell_cols("A:V"), col_names = TRUE, col_types = NULL) # load excel file 
```

```r
alum <- alum %>% 
    clean_names() 
```

```r
head(alum)
```

```r
alum$first_name <- gsub('[\\S\\s]*?',"",alum$first_name)
# removed preffered name to just legal name for matching purposes 
```

Bring in data for applicants

```r
alum %>%
    tabyl(gender)

alum <- alum %>% mutate(gender=recode(gender, 
                         `M`=1,
                         `F`=0))

alum %>%
    tabyl(gender)
```

```r
alum$still_in_high_school[is.na(alum$still_in_high_school)] <- 0

alum <- alum %>% mutate(still_in_high_school=recode(still_in_high_school, 
                         `High School`=1,
                         `0`=0))


alum %>%
    tabyl(still_in_high_school)
```

```r
alum %>%
    tabyl(upci_participation_year)

alum <- alum %>% 
    separate(upci_participation_year, c("year_a","year_b","year_c","year_d"), sep = "[/,]", remove = FALSE, extra = 'merge', fill = "right") 
```
